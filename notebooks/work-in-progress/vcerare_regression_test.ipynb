## Compare VCERARE outputs without blowing up memory

```python
import polars as pl
from polars.testing import assert_frame_equal
from pudl.workspace.setup import PudlPaths

import duckdb
```

```python
for year in [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]:
    print(year)
    assert_frame_equal(
        pl.scan_parquet(PudlPaths().parquet_path("out_vcerare__hourly_available_capacity_factor")).filter(pl.col("report_year") == year),
        pl.scan_parquet("https://s3.us-west-2.amazonaws.com/pudl.catalyst.coop/nightly/out_vcerare__hourly_available_capacity_factor.parquet").filter(pl.col("report_year") == year),
        check_column_order=False,
        check_dtypes=False,
    )
```
