```python
s = 'Turbines, Inverters, or Hydrokinetic Buoys'
s = re.sub(r'[^0-9a-zA-Z]+', ' ', s)
s = s.strip().lower()
s = s.replace(' ', '_')
s
```

```python
def split_generators(gen_string):
    split_g = gen_string.split(',')
    split_g = [re.sub(r'[^0-9a-zA-Z]+', ' ', x) for x in split_g]
    split_g = [x.lower().strip() for x in split_g]
    split_g = [x.replace(' ', '_') for x in split_g]
    return split_g
```
